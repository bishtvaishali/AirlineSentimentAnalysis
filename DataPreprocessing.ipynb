```python
# tokenization, stemming, stop word removal

import xlrd, re
path = ("Tweets.xls") 
spreadSheet = xlrd.open_workbook(path)
sheet = spreadSheet.sheet_by_index(1) 

testData = []
tweetsData = []
sentiment = [] 

#Training Data ------
for i in range(1, 270):
    arrStr = re.sub('[!@#\"$]', "", sheet.cell_value(i, 10)) 
    arrStr = re.sub(r"\//\S+", "", arrStr)
    tweetsData.append(arrStr) 
     
# print(tweetsData)




#stemmer = PorterStemmer() // stemming has to "ha"
 
#data = "!! The: ; , , had having has it's at the All work and no the the play makes jack dull boy."

#stemmer=SnowballStemmer("english", ignore_stopwords= True)
# import nltk

from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

stemmer=SnowballStemmer("english") # changng "opportunity" to opportun 
stopWordList = set(stopwords.words('english'))

tokenList=[]


for data in tweetsData:
    wlist = word_tokenize(data)
    tokenList.extend(wlist)
 
punctuations="?:!.,;...@"
#lemmitization

wordList = []
for token in tokenList:
   # print('token:', token)
    token= stemmer.stem(token.lower())
    token = re.sub(r"http\S+", "", token) #removing url from token
    token = re.sub(r"\//\S+", "", token) #removing url from token  for this: //t.co/mWpG7grEZP
   # print(token)    
    if token not in stopWordList and token not in punctuations:
        wordList.append(token)

    
print(wordList) 


```
